# Pre-processing

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

# Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

```python
compare = []
for i in range(len(y_test)):
    temp = [0,0]
    temp[0] = np.array(y_test)[i]
    temp[1] = y_pred[i]
    compare.append(temp)
pd.DataFrame(compare,columns = ["origianl","Predicted"]).head(10)
```

```python
compare_graph = []
correct = 0
incorrect = 0
for i in range(len(y_test)):
    if(np.array(y_test)[i]==y_pred[i]):
        compare_graph.append(1)
        correct += 1
    else:
        compare_graph.append(0)
        incorrect += 1
# corr_incorr = pd.DataFrame([[correct,incorrect]],columns = ["Correct","In correct"])
fig,ax = plt.subplots()
ax.plot(compare_graph)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(["Correct","In correct"],[correct,incorrect],color = ["g","r"])
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm)
```

```python
print("accuracy : ", accuracy_score(y_test,y_pred)*100 ,"%")
```
